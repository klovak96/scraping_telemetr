```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
df = pd.read_csv('results.csv')
df.tail()
```

```python
def get_tg_admins(links_df):
    
    links = links = df['link'].values
    links = [link.replace('@', '') for link in links]
    base_url = 'https://telegram.me'
    descriptions=[]
    all_admins = []
    sub_count = []
    i = 0
    for link in links:
        i += 1
        print(i)
        response = requests.get(f'{base_url}/{link}')
        soup = BeautifulSoup(response.text)
        try:
            desription = soup.find('div', 'tgme_page_description').text
            descriptions.append(desription)
        except:
            descriptions.append('Null')
        
        admins = []
        
        try:
            a = soup.find('div', 'tgme_page_description').find_all('a')
            for link in a:
                admin = link.get('href')
                if re.search(r'^https://t\.me/(?!\+)(?!.*[bB][oO][tT]$).*', admin):
                    admins.append(admin)
            if len(admins) == 0:
                admins.append('Null')
        except:
            admins.append('Null')
    
        all_admins.append('\n'.join(admins))
        
        try:
            sub = soup.find('div', 'tgme_page_extra').text
            sub = re.sub('\D', '', sub)
            sub_count.append(sub)
        except:
            sub_count.append('Null')
            
        
        
    return sub_count, descriptions, all_admins

df['sub'], df['descriptions'], df['admins'] = get_tg_admins(df['link'])
```
