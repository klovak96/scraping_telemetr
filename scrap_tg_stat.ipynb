```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
df = pd.read_csv('results.csv')
df.tail()
```

```python
df = df.head(15)
```

```python
def get_tg_admins(links_df):
    
    links = links = df['link'].values
    links = [link.replace('@', '') for link in links]
    base_url = 'https://tgstat.ru/channel'
    last_posts=[]

    i = 0
    for link in links:
        i += 1
        print(i)
        response = requests.get(f'{base_url}/{link}')
        soup = BeautifulSoup(response.text)
        print(soup)
        try:
            last_post = soup.find('div', 'card card-body border p-2 px-1 px-sm-3 post-container').find('small').text
            last_posts.append(last_post)
        except:
            last_posts.append('Null')
        
        
        
    return last_posts

df['last_post'] = get_tg_admins(df['link'])
```
